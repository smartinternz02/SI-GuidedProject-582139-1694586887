### Importing the necessary libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Taking the 'car_crashes' dataset

```python
print(sns.get_dataset_names())
```

### Loading the dataset

```python
df = sns.load_dataset('car_crashes')
df
```

First, let us know about the details of the dataset.

```python
df.info()
```

We see that there are no null values and there are 8 columns in the dataset. By observing the dataset we can see that it consists of information about car crashes in all 51 states of the United States. The dataset holds columns which represents parameters pertaining to the percentage of car crashes based on the reason and the insurance details.

## Visualizing the data

### Bar plot

This plot is to observe the insurance premium in each state.

```python
sns.barplot(x = "abbrev", y = "ins_premium", data = df)
plt.xticks(rotation= 90)
plt.show()
```

Here, we can see which states offer the highest and lowest insurance premium.

### Correlation and Heatmap

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, annot = True)
```

The heatmap allows us to see the correlation of all the variables with each other. Light coloured boxes indicate high correlation. Each square shows the correlation between the variables on each axis. Correlation ranges from -1 to +1. Values closer to zero means there is no linear trend between the two variables. The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar, but instead of both increasing one variable will decrease as the other increases. The diagonals are all 1/white because those squares are correlating each variable to itself (so it's a perfect correlation). For the rest the larger the number and lighter the color the higher the correlation between the two variables. The plot is also symmetrical about the diagonal since the same two variables are being paired together in those squares.

### Scatterplot

This plot is to represent the insurance premium vs. insurance losses incurred by the insurance company.

```python
sns.scatterplot(y = "ins_losses", x = "ins_premium", data = df)
```

We can see that the insurance losses roughly increase with increase in the insurance premium.

### Relationship plot

```python
sns.relplot(y = "ins_losses", x = "ins_premium", data = df, hue = "abbrev")
```

With relationship plot we can see the relation between insurance premium and losses aling with the states which offer it.

### Joint plot

```python
sns.jointplot(x = "ins_premium", y = "ins_losses", data = df)
```

Joint plot gives us the distribution of losses and insurnce individually and also how both are related.

### Distribution plot

```python
sns.distplot(df["speeding"])
```

This distribution plot gives us an idea about how the variable speeding is distributed.
